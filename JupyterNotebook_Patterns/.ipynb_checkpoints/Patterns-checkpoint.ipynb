```java
/*
5> Print The Following:

         *
       * *
     * * *
   * * * *
 * * * * *

*/

System.out.print("Enter the Enter:");
Scanner sc = new Scanner(System.in);
int n = sc.nextInt();
System.out.println("Pattern_5");



for (int i = 1; i <=n ; i++){
    
    for(int k = 1; k <= (n-i); k++){
        System.out.print(" ");
    }
    
    for(int j = 1; j <= i; j++){
        System.out.print("*");
    }
    
    System.out.println();
}

```
